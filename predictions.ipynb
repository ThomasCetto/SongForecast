```python
import py_midicsv as pm

# Load the MIDI file and parse it into CSV format
csv_string = pm.midi_to_csv("midiFiles/track1.midi")
csv_string = ["Track, Time, Name, Channel, Note, Velocity\n"] + csv_string

with open("converted1.csv", "w") as f:
    f.writelines(csv_string)
```

```python
# ogni singola nota ha:
# numero traccia(1), tempo, Note_on_C, canale(0), nota, velocità (intensità da 0 a 127, 127 è forte)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('converted1.csv', sep=r'\s*,\s*')
```

```python
df
```

```python

df_notes = df.drop(df[df.Velocity == 0].index)
df_notes = df_notes.drop(df[(df.Name != "Note_on_c") & (df.Name != "Control_c")].index)
df_notes
```

```python
all_notes = df_notes["Note"].tolist()
last_note = [70] + all_notes[:len(all_notes)-1]
all_vel = df_notes["Velocity"].tolist()
last_vel = [0] + all_vel[:len(all_vel)-1]


df_notes["1NoteBefore"] = last_note
df_notes["1VelocityBefore"] = last_vel
```

```python
df_notes
```

```python
#Cancella la prima riga
df_notes.drop(index=df.index[0], 
        axis=0, 
        inplace=True
)
```

```python
df_notes.to_csv('out.csv')
```

```python
# Parse the CSV output of the previous command back into a MIDI file
midi_object = pm.csv_to_midi("converted1.csv")

# Save the parsed MIDI file to disk
with open("example_converted.mid", "wb") as output_file:
    midi_writer = pm.FileWriter(output_file)
    midi_writer.write(midi_object)
```
